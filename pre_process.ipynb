```python
import pandas as pd
from urllib.parse import urlparse
```

```python
"""
Raw Data 구조

Id와 Query String에는 중복된 값이 없음
PopularityRank는 검색어 횟수의 빈도가 많은 순서
PredictedLabel은 해당 검색이 Code Search의 의도가 있었는지 판별하는 Flag로 정확도는 약72%
"""

df_java = pd.read_csv('./java.csv')
df_java['Language'] = 'Java'
df_java
```

```python
df_c_sharp = pd.read_csv('./c#.csv')
df_c_sharp['Language'] = 'C#'
df_c_sharp
```

```python
df = pd.concat([df_java, df_c_sharp], ignore_index=True)
df
```

```python
"""
TopClickedUrls는 3개의 URL로 구성되어 있고, Query-Url Mapping을 위해서는 단일 URL로 분리가 필요하다
"""

df.iloc[0, df.columns.get_loc('TopClickedUrls')]
```

```python
"""
데이터 전처리 진행

query: QueryString
url : url의 http(s)://와 www. 삭제 (e.g., www.java.com과 java.com은 같은 URL로 처리하기 위함) 
domain: url의 domain (e.g., https://www.java.com)
rank: PopularityRank
code_search: PredictedLabel 값을 String type의 TRUE, FALSE로 저장
"""

new_data = {
    'query_id': [],
    'url_id': [],
    'domain_id': [],
    'query': [],
    'url': [],
    'origin_url': [],
    'domain': [],
    'rank': [],
    'code_search': [],
    'language': [],
}

url_size = 0
url_dict = {}
url_dict_count = {}

domain_size = 0
domain_dict = {}
domain_dict_count = {}

for idx in df.index:
    for origin_url in df.iloc[idx, df.columns.get_loc('TopClickedUrls')].split(','):
        domain = urlparse(origin_url).netloc
        if not domain:
            print('[Invalid Data]', origin_url)
            # 잘못된 형식의 URL 데이터 필터링 진행
            continue

        url = origin_url.split('://www.')[-1]
        url = url.split('://')[-1]
        
        if url not in url_dict:
            url_dict[url] = url_size
            url_dict_count[url] = 0
            url_size += 1
        url_id = url_dict[url]
        url_dict_count[url] += 1

        if domain not in domain_dict:
            domain_dict[domain] = domain_size
            domain_dict_count[domain] = 0
            domain_size += 1
        domain_id = domain_dict[domain]
        domain_dict_count[domain] += 1

        new_data['query_id'].append(idx)
        new_data['query'].append(df.iloc[idx, df.columns.get_loc('QueryString')])
        new_data['origin_url'].append(origin_url)
        new_data['url_id'].append(url_id)
        new_data['url'].append(url)
        new_data['domain_id'].append(domain_id)
        new_data['domain'].append(domain)
        new_data['rank'].append(df.iloc[idx, df.columns.get_loc('PopularityRank')])
        new_data['code_search'].append('TRUE' if df.iloc[idx, df.columns.get_loc('PredictedLabel')] else 'FALSE')
        new_data['language'].append(df.iloc[idx, df.columns.get_loc('Language')])
df = pd.DataFrame(new_data)
df.index.name = 'id'
```

```python
df.to_csv('./data.csv')
```
